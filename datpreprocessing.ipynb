```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies[['id', 'genres', 'keywords', 'title', 'tagline', 'overview', 'production_companies']]
```

```python
credits = credits[['movie_id', 'cast', 'crew']]
```

```python
credits
```

```python
#Lets join/merge both df based on id - movie_id
df = pd.merge(movies, credits, left_on='id', right_on='movie_id', how='inner')
```

```python
df
```

```python
df = df.drop(columns = ['movie_id'], axis = 1)
```

```python
import ast
def get_names(desc):
    list_items = ast.literal_eval(desc)
    names = []
    for item in list_items:
        names.append(item['name'])
    comma_desc = ','.join(names)
    return comma_desc
```

```python
df['genres'] = df['genres'].apply(get_names)
```

```python
df['keywords'] = df['keywords'].apply(get_names)
```

```python
df['production_companies'] = df['production_companies'].apply(get_names)
```

```python
df['cast'] = df['cast'].apply(get_names)
```

```python
df['crew'] = df['crew'].apply(get_names)
```

```python
df.sample()
```

```python
df.to_csv('cleaned_data.csv')
```

```python
df.fillna('', inplace = True)
```

```python
#Lets concatenate all columns except id, title
df['tags'] = df['genres'] + ',' +  df['keywords'] + ',' + df['tagline'] + ',' + df['overview'] + ',' + df['production_companies'] + ',' + df['cast'] + ',' + df['crew']
```

```python
df = df[['id', 'title', 'tags']]
```

```python
df['tags'] = df['tags'].str.lower()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Download stopwords if not already available
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
stop_words = set(stopwords.words('english'))
def remove_stop_words(all_words):
    all_words = all_words.replace('.', '')
    all_words = all_words.replace(' ', ',')
    all_words = all_words.split(',')
    standard_words = []
    for word in all_words:
        word = word.strip()
        if word != '':
            if word not in stop_words:
                if word not in standard_words:
                    standard_words.append(word)
    tags = ','.join(standard_words)
    return tags
```

```python
df['tags'] = df['tags'].apply(remove_stop_words)
```
