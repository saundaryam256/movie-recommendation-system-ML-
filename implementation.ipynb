```python
import pandas as pd 
import numpy as np 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('cleaned_dataa.csv')
```

```python
df.info()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#CountVectorizer is a tool in scikit-learn that converts text data into numbers.
cv=CountVectorizer(max_features=10000, stop_words='english')
#Use only the top 10,000 most frequent words from the dataset.
#Remove common English words that do not add meaning.
```

```python
vector= cv.fit_transform(df['tags']) # analyze the vocaboury and make sparse matrix of the it 
```

```python
vector=vector.toarray() # convert sparsr matrix to the simple array
```

```python
cv.get_feature_names_out() 
```

```python
len(df.loc[0, 'tags'].split(','))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
#Lets write function get name of movie by index
def get_name_by_index(i):
    if i < len(df) and i>0:
        return df.loc[i,'title']
    else:
        return''
```

```python
def get_index_from_name(name):
    findex=-1
    for i in df. index:
        if df.loc[i,'title'].strip().lower()==name.strip().lower():
            findex=i
            break
    return findex
```

```python
name = input("Enter movie you watched:")
index = get_index_from_name(name)
if index != -1:
    similarity_indexes =  list(enumerate(similarity[index]))
    similarity_indexes = sorted(similarity_indexes, key=lambda x: x[1], reverse=True)
    print("You must watch following movies:")
    for i in range(1, 6):
        print(i, ":", get_name_by_index(similarity_indexes[i][0]))
else:
    print("Movie Not Found")    
```

```python
#Lets export similarity 
import pickle as pkl
pkl.dump(similarity, open('similarity.pkl', 'wb'))
```
